We use the metric distance $\rho_{i,j} = \sqrt{2 - (1 - \rho_{i,j})}$.  Stocks with returns that move closely together will have high correlation and small distances.

The dendrogram method works by linking the vertices into clusters repeatedly  starting with the closest vertices.

# Sequential Portfolio Selection Problem

The sequential portfolio selection problem works as follows:

1. Given historical returns of each asset
2. Filter to select K assets
3. For each time step:
    *  Choose a portfolio consisting of some amount of weights assigned to each of the K ossets.
    *  Observe the returns; we receive a weighted average of the rewards in question.
    *  Update accumulated knowledge so that we can choose the portfolio appropriately in the next time step.


The paper describes a method in section 2.3 that selects stocks by first choosing each asset one time, and then chooses the stock with the highest value for the upper confidence bound of empircal average return.

It also describes another way to get weights based on selecting a portfolio to mimnimize conditional value at risk.  It suggests taking a convex combination of the single-asset portfolio and the risk-aware portfolio to acheive a desired return between reward and risk.

They define the upper confidence bound as:
$$
\begin{equation}
\arg\underset{a}{\max} \left\{{\tilde R}_t(a) + c \sqrt{2 \frac{\ln t}{N_t(a)}}\right\}
\end{equation}
$$

They calculate the weights at time $t$ by selecting a positive vector of weights ${\bf u}$ summing to $1$ and a real number $\alpha$ so as to minimize

$\frac{1}{(\delta + t - 1)(1-\gamma)} \left( \sum_{s=1}^\delta -{\bf u} ^{T} {\bf H}_s - \alpha + \sum_{s=1}^{t-1} - {\bf u}^{T} {\bf R}_s - \alpha \right)$

Where $H_s$ is the $i$th historical return and $R_s$ is the $i$th vector of returns for the trading period.


# Upper Confidence Bound

To use an upper-confidence bound, we add a boost to actions that that have not been explored as much; thus we can prioritize exploring actions that we are uncertain about:
$$
\begin{equation}
A_t = \arg\underset{a}{\max} \left\{Q_t(a) + c \sqrt{\frac{\ln t}{N_t(a)}}\right\}
\end{equation}
$$

To use the upper-confidence bound in a multi-armed bandit, we have a repeated loop:

* Select an action using the upper-confidence Bound $A_t$
* Update $Q_t$ based on the result of the action (reward).
* Update $A_t$ based on the new Q_t using the new counts.

The upper-confidence bound (UCB) algorithm maintains a sample mean reward and confidence interval for each action. On each round, it selects the action with the highest upper confidence bound of its reward. This strategy optimistically favors arms with high sample means and high uncertainties. As arms are sampled over time, their uncertainties shrink and low-performing arms are eliminated. A tuning parameter controls the width of the confidence bounds. Wider bounds encourage more initial exploration before converging on the empirically best arm.



````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import rand, seed


def optimal_action(qvalue, eps):
    """
    Determines what is the action to take given a measure of past
    expected rewards across actions. With probability eps the action
    is not the greedy one
    """
    nactions = qvalue.shape[0]
    action_hat = np.where(qvalue == np.max(qvalue))

    if rand() <= eps:
        randnum = rand()
        for aa in range(nactions):
            if randnum < (aa + 1) / nactions:
                break
    elif action_hat[0].shape[0] > 1:
        # Randomize action when ties
        randnum = rand()
        for aa in range(action_hat[0].shape[0]):
            if randnum < (aa + 1) / action_hat[0].shape[0]:
                break
        aa = action_hat[0][aa]
    else:
        aa = np.argmax(qvalue)

    return aa
````

````python
def reward_update(action, reward, qvalue_old, alpha):
    qvalue_new = qvalue_old.copy()

    qvalue_new[action] = qvalue_old[action] + alpha * (reward - qvalue_old[action])

    return qvalue_new
````

````python
# Bandit problem for stock selection

NK = 29
EPSILON = 0.00
ALPHA = 0.975
NEPISODES = 1000
HOLD = 1
TMAX = YData.shape[0] - HOLD

seed(1234)

reward_avg = np.zeros((TMAX))
optimal_avg = np.zeros((TMAX))

reward_queue = np.zeros((HOLD, 2))

for run in range(NEPISODES):
    # Initialize q function and actions record
    qvalue = np.zeros((NK))
    qvalue_up = np.zeros((NK))
    nactions = np.zeros((NK))
    for tt in range(TMAX):
        aa_opt = optimal_action(qvalue_up, 0.00)
        nactions[aa_opt] += 1
        # Compute reward as return over holding period
        reward_queue[HOLD - 1, 0] = (
            YData[tt + HOLD, aa_opt] - YData[tt, aa_opt]
        ) / YData[tt, aa_opt]
        reward_queue[HOLD - 1, 1] = aa_opt
        # Update Q function using action chosen HOLD days before
        qvalue = reward_update(
            int(reward_queue[0, 1]), reward_queue[0, 0], qvalue, ALPHA
        )
        # Upper-confidence adjustment
        qvalue_up = np.zeros((NK))
        for aa in range(NK):
            if nactions[aa] == 0:
                qvalue_up[aa] = np.max(qvalue) + 1.0
            else:
                qvalue_up[aa] = qvalue[aa] + np.sqrt(np.log(tt + 1) / nactions[aa])
        reward_queue[0 : HOLD - 1, :] = reward_queue[1:HOLD, :]
        reward_avg[tt] += reward_queue[HOLD - 1, 0] / NEPISODES
        optimal_avg[tt] += (
            aa_opt == np.argmax((YData[tt + HOLD, :] - YData[tt, :]) / YData[tt, :])
        ) / NEPISODES
````

````python
plt.plot(data2.index[HOLD : YData.shape[0]], optimal_avg)
plt.title("Frequency of optimal action", fontsize="x-large")

fig = plt.gcf()
fig.set_size_inches(18, 3)
plt.show()
````

````python
# Average annualized return from holding the Bandit portfolio
print(
    (1 + np.mean(reward_avg)) ** (250 / HOLD) - 1,
    np.sqrt(250 / HOLD) * np.std(reward_avg),
)
````

````python
return_cumulative = np.zeros((TMAX + 1, 2))
return_cumulative[0, 0] = 1
return_cumulative[0, 1] = 1
for tt in range(1, TMAX + 1):
    return_cumulative[tt, 0] = return_cumulative[tt - 1, 0] * (1 + reward_avg[tt - 1])
    rmean = np.mean(
        (YData[tt + HOLD - 1, :] - YData[tt - 1, :]) / YData[tt - 1, :]
    )
    return_cumulative[tt, 1] = return_cumulative[tt - 1, 1] * (1 + rmean)

plt.plot(
    data2.index[HOLD - 1 : YData.shape[0]],
    return_cumulative[:, 0],
    label="Cumulative return Bandit",
)
plt.plot(
    data2.index[HOLD - 1 : YData.shape[0]],
    return_cumulative[:, 1],
    label="Cumulative return Average",
)  # noQA E203
legend = plt.legend(loc="upper left", shadow=True)
plt.title("Cumulative Return using Upper Confidence Bound Algorithm")
plt.ylabel("Portfolio Value as percentage of initial value")
plt.xlabel("Date")
plt.xticks(rotation=70);
````

# Epsilon-Greedy

The epsilon-greedy algorithm is a simple and widely used approach for balancing exploration and exploitation in reinforcement learning and multi-armed bandit problems.

With probability $(1-\epsilon)$ we choose the action which currently has the highest expected reward.  With probability $\epsilon$ we choose an action uniformly at random.


```
# Initialize variables
epsilon = 0.1  # The exploration rate (0.0 to 1.0)
num_actions = number of available actions
action_values = array of length num_actions, initialized to 0
action_counts = array of length num_actions, initialized to 0

# Function to select an action based on epsilon-greedy policy
function choose_action():
    if random_number() < epsilon:
        # Explore: Choose a random action with probability epsilon
        selected_action = random_action()
    else:
        # Exploit: Choose the action with the highest estimated value
        selected_action = argmax(action_values)
    return selected_action

# Function to update action-value estimates after taking an action
function update_action_value(action, reward):
    action_counts[action] += 1
    n = action_counts[action]  # Number of times action has been chosen so far
    current_value = action_values[action]
    updated_value = current_value + (1 / n) * (reward - current_value)
    action_values[action] = updated_value

# Main loop for the agent
for each time step:
    # Choose an action based on epsilon-greedy policy
    selected_action = choose_action()

    # Simulate taking the selected action and receiving a reward
    reward = environment.simulate_action(selected_action)

    # Update action-value estimate based on the received reward
    update_action_value(selected_action, reward)
```

First, we try using a multi-armed bandit algorithm using an upper confidence bound approach to select stocks over a period of time. It defines a multi-armed bandit problem with NK=29 arms, representing 29 different stocks to choose from.
At each timestep, an optimal stock is chosen based on the upper confidence bound of the q-values. This balances exploiting the best stock so far with exploring uncertain options.
Rewards are calculated as the return over a holding period of HOLD days. This measures performance over a multi-day investment horizon.
After each holding period, the q-values are updated using the stock chosen 2 days ago, its reward, and an alpha learning rate.
The upper confidence bounds are adjusted using the number of times each stock was chosen. More choices lower the bound.
The average reward and optimal selections are tracked to compare the algorithm's performance to an optimal strategy.
Overall, this implements a reinforcement learning approach to dynamically select stocks over time. The use of upper confidence bounds balances exploiting current knowledge with continued exploration. The holding periods and multiple episodes allow evaluating performance realistically across a long timespan with many market conditions.

````python
# Bandit problem for stock selection
import numpy as np

NK = 29
EPSILON = 0.01
ALPHA = 0.95
NEPISODES = 1000
HOLD = 2
TMAX = YData2.shape[0] - HOLD

np.random.seed(1234)

reward_avg = np.zeros((TMAX))
optimal_avg = np.zeros((TMAX))

reward_queue = np.zeros((HOLD, 2))

for run in range(NEPISODES):
    # Initialize q function and actions record
    qvalue = np.zeros((NK))
    qvalue_up = np.zeros((NK))
    nactions = np.zeros((NK))
    for tt in range(TMAX):
        aa_opt = optimal_action(qvalue_up, EPSILON)
        nactions[aa_opt] += 1
        # Compute reward as return over holding period
        reward_queue[HOLD - 1, 0] = (
            YData2[tt + HOLD, aa_opt] - YData2[tt, aa_opt]
        ) / YData2[tt, aa_opt]
        reward_queue[HOLD - 1, 1] = aa_opt
        # Update Q function using action chosen HOLD days before
        qvalue = reward_update(
            int(reward_queue[0, 1]), reward_queue[0, 0], qvalue, ALPHA
        )
        # Upper-confidence adjustment
        qvalue_up = np.zeros((NK))
        for aa in range(NK):
            if nactions[aa] == 0:
                qvalue_up[aa] = np.max(qvalue) + 1.0
            else:
                qvalue_up[aa] = qvalue[aa] + np.sqrt(np.log(tt + 1) / nactions[aa])
        reward_queue[0 : HOLD - 1, :] = reward_queue[1:HOLD, :]
        reward_avg[tt] += reward_queue[HOLD - 1, 0] / NEPISODES
        optimal_avg[tt] += (
            aa_opt == np.argmax((YData2[tt + HOLD, :] - YData2[tt, :]) / YData2[tt, :])
        ) / NEPISODES
````

Here is a plot of the cumulative returns:

````python
return_cumulative = np.zeros((TMAX + 1, 2))
return_cumulative[0, 0] = 1
return_cumulative[0, 1] = 1
for tt in range(1, TMAX + 1):
    return_cumulative[tt, 0] = return_cumulative[tt - 1, 0] * (1 + reward_avg[tt - 1])
    rmean = np.mean(
        (YData2[tt + HOLD - 1, :] - YData2[tt - 1, :]) / YData2[tt - 1, :]
    )
    return_cumulative[tt, 1] = return_cumulative[tt - 1, 1] * (1 + rmean)

plt.plot(
    data3.index[HOLD - 1 : YData2.shape[0]],
    return_cumulative[:, 0],
    label="Cumulative return Bandit",
)
plt.plot(
    data3.index[HOLD - 1 : YData2.shape[0]],
    return_cumulative[:, 1],
    label="Cumulative return Average",
)  # noQA E203
legend = plt.legend(loc="upper left", shadow=True)
plt.title("Cumulative Return using Upper Confidence Bound Algorithm")
plt.ylabel("Portfolio Value as percentage of initial value")
plt.xlabel("Date")
plt.xticks(rotation=70);
````

With the equally-weighted portfolio we end up with 78% of our original value, which means we lose 22%, while the $\epsilon$-greedy algorithm gives us 76% of the original value, which means that we lose 24%.

However, this better performance does not take into account any additional risk from only holding a single stock at a time as opposed to a diversified portfolio of 30 stocks.

Since the upper-confidence bound algorithm performed better, it appears that the use of uncertainty information to guide structured exploration was effective. It seeks out potentially high-reward arms that need more samples to confidently estimate their qualities. In contrast, $\epsilon$-greedy performs uninformed random exploration by selecting arms uniformly. UCB focuses sampling on informative arms, while epsilon-greedy continues to devote samples to exploring suboptimal arms. These differences lead UCB to converge more quickly and reliably on the optimal arm compared to epsilon-greedy given the same number of samples.

This observation raises questions about the effectiveness of these $\epsilon$-greedy strategy in such a volatile market environment. One might expect the volatility to mean that the upper-confidence bound algorithm might learn incorrect information early on, but in fact $\epsilon$-greedy performed worse.  It suggests that we are not able to exploit potential market trends or identify high-performing stocks effectively during this specific time frame.  The upper-confidence algorithm seems like it is selecting a better portfolio by not making as many random choices. The Upper-Confidence Bound portfolio algorithm always invests in the asset that maximizes the estimated upper confidence bound of return for a given confidence level.  On the other hand, the $\epsilon$-greedy algorithm always invests in a single asset; with probability $1-\epsilon$ this is the one with the best estimated return so far, and with probability $\epsilon$ it invests in one of the stocks at random.
