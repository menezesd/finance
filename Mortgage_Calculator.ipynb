```python
import numpy_financial as npf
```

```python
def print_amortization_table(principal, rate, years):
    monthly_rate = rate/12
    periods = years * 12
    pmt = npf.pmt(monthly_rate, periods, -principal, 0)
    bal = principal
    total_interest = 0
    print("Time  Unpaid Principal Balance\tInterest\tPayment\tTotal Interest Paid\tNew Balance")
    for i in range(periods):
      total_interest += bal * monthly_rate
      new_bal = bal + bal * monthly_rate - pmt
      print (f"{i}\t${bal:-10.2f}\t\t${bal * monthly_rate:-8.2f}\t${pmt:-5.2f}\t${total_interest:-10.2f}\t${new_bal:-10.2f}")
      bal = new_bal
```

```python
print("Amortization Table for 30-year fixed rate mortgage at 4.0% per annum")
print_amortization_table(1000000, 0.04, 30)
```

```python
print("Amortization Table for 20-year fixed rate mortgage at 2.5% per annum")
print_amortization_table(1000000, 0.025, 20)
```

```python
def print_amortization_table_variable_rate(principal, rates, years):
    periods = years * 12

    bal = principal
    total_interest = 0
    print("Time  Unpaid Principal Balance\tInterest\tPayment\tTotal Interest Paid\tNew Balance")
    for i in range(periods):
      monthly_rate = rates[i//12] / 12 / 100
      pmt = npf.pmt(monthly_rate, periods-i, -bal, 0)
      total_interest += bal * monthly_rate
      new_bal = bal + bal * monthly_rate - pmt
      print (f"{i}\t${bal:-10.2f}\t\t${bal * monthly_rate:-8.2f}\t${pmt:-5.2f}\t${total_interest:-10.2f}\t${new_bal:-10.2f}")
      bal = new_bal
```

```python
print_amortization_table_variable_rate(10**6, [8.35]*7 + [6.83, 8.06, 7.13, 7.16, 5.93, 5.85, 5.81, 6.22, 6.18, 6.17, 5.10, 5.14, 4.86, 3.95, 3.35, 4.48, 3.87, 4.01, 4.29, 3.99, 4.55, 3.74, 2.67], 30)
```
